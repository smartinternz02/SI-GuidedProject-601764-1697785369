```python
plt.figure(figsize=(5, 5))
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age(Years)'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Level of Education', order=df['Level of Education'].value_counts().index)
plt.title('Level of Education Distribution')
plt.xlabel('Count')
plt.ylabel('Level of Education')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Are you involved in any sports?')
plt.title('Involvement in Sports')
plt.xlabel('Are you involved in any sports?')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Performance in online', y='Study time (Hours)')
plt.title('Study Time vs. Performance')
plt.xlabel('Performance in online')
plt.ylabel('Study time (Hours)')
plt.show()
```

```python
numeric_columns = ['Age(Years)', 'Study time (Hours)', 'Performance in online', 'Clearing doubts with faculties in online mode']
sns.pairplot(df[numeric_columns])
plt.suptitle('Pairplot of Numeric Variables', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Internet facility in your locality')
plt.title('Internet Facility in Locality')
plt.xlabel('Internet Facility')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='Economic status', order=df['Economic status'].value_counts().index)
plt.title('Economic Status Distribution')
plt.xlabel('Count')
plt.ylabel('Economic Status')
plt.show()
```
